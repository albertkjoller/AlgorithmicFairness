```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('seaborn-v0_8-pastel')
```

```python
df = pd.read_csv('../../data/catalan-juvenile-recidivism-subset.csv')
```

```python
df.head()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, gridspec_kw={'width_ratios': [1.5,1,1],}, figsize=(12,6))
fig.suptitle('Recidivism based on Sex')
df['V1_sex'].value_counts().plot.pie(title='Sex Distribution',ylabel=' ',autopct='%1.1f%%', ax=axes[0])
df.loc[df['V1_sex'] == 'male', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Male Recidivism',ylabel=' ',autopct='%1.1f%%',labels=['No', 'Yes'],ax=axes[1])
df.loc[df['V1_sex'] == 'female', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Female Recidivism',ylabel=' ',autopct='%1.1f%%',labels=['No', 'Yes'],ax=axes[2])
```

```python
df['V115_RECID2015_recid'].value_counts().plot.pie(title='Recidivism',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'])
```

```python
df['V4_area_origin'].value_counts().plot.pie(title='Area of Origin',ylabel=' ',autopct='%1.1f%%')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, gridspec_kw={'width_ratios': [3,1,1], 'height_ratios': [2,1.5]}, figsize=(12,6))
fig.suptitle('Recidivism based on Area of Origin')
df['V4_area_origin'].value_counts().plot.pie(title='Area of Origin Distribution',ylabel=' ',autopct='%1.1f%%', ax=axes[0,0])
df.loc[df['V4_area_origin'] == 'Maghreb', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Maghreb',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'],ax=axes[0,1])
df.loc[df['V4_area_origin'] == 'Europe', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Europe',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'],ax=axes[0,2])
df.loc[df['V4_area_origin'] == 'Latin America', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Latin America',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'],ax=axes[1,0])
df.loc[df['V4_area_origin'] == 'Spain', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Spain',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'],ax=axes[1,1])
df.loc[df['V4_area_origin'] == 'Other', 'V115_RECID2015_recid'].value_counts().plot.pie(title='Other',ylabel=' ',autopct='%1.1f%%', labels=['No', 'Yes'],ax=axes[1,2])
```
